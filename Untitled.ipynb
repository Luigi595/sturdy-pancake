```python
import webbrowser
import os
import pandas as pd
```

```python
#Data collection
#A very rudimentary approach to obtain the data; this opens a lot of browser tabs in order to download the files
data= ["094","060","061","063","053","055","610","051","613","095","096","058","054","801","093","056","091","617","601","099","020","009","600","700","015","001","002","003","004","006","007","008","200","400","100","300","500","010","011","012","016","013","014","019","005","604","607","064","065","066","092","068","067","069","090","071","070","072","073","074","075","076","077","062","078","079","082","083","084","085","086","087","097","080","081","052","088","059","089","098","021","022","023","024","025","026","027","028","029","031","032","034","035","033","000","998","030"]
for i in data:
    webbrowser.open_new_tab("http://www.estadistica.unam.mx/reportesinstitucionales/reporte_pobxcarrera.php?cve_dep="+i+"&enviar=Excel")
```

```python
#Parsing the data
#Having downloaded the data, and then moving the files to a folder "data" in the same directory as this notebook
#Running this cell ignores files smaller than 1kb and obtains the last row of every table with records from 2017-2018
comp=[]
for filename in os.listdir("data"):
    if os.path.getsize("data/"+filename)>1000:
        df=pd.read_html("data/"+filename)
        for i in df:
            if i.iloc[-1][1]=='2017-2018' and not i.iloc[-1][4]=="0":
                comp.append(i.iloc[-1])
```

```python
#Starting the dataframe and converting the column tipes to int
ndf=pd.DataFrame(comp)
ndf=ndf.astype({0:int,2:int,3:int,4:int,5:int,6:int,7:int,8:int})
```

```python
#Adding together the values for the same career and displaying the dataframe so far
ndf=ndf.groupby([0]).sum()
ndf
```

```python
#Calculates the propotion of men that are freshmen, of those who are not and the difference between those proportions
phombres=ndf[2]/ndf[4]
rhombres=ndf[5]/ndf[7]

calc=rhombres-phombres
```

```python
#Generates a new dataframe with our interest values
output=pd.DataFrame({"Primer Ingreso":phombres,"Reingreso":rhombres,"Diferencia":calc})
```

```python
#Saves the dataframe to a csv file
output.to_csv("output.csv")
```
